Preprocessing Data,
Remove ID (not a relevant predictor)

Hemisphere is also removed because continent is a more accurate predictor. Maybe this is a mistake (to be checked)

```python
df = df.drop(['Patient ID','Hemisphere'], axis = 'columns')
df.head()
```

Converting blood pressure to a single variable (Mean Arterial Bloodpressure = MAP) 
https://www.researchgate.net/post/Is_there_any_way_to_convert_systolic_and_diastolic_blood_pressure_into_a_single_indicator_variable

```python
bloodpress = list(df['Blood Pressure'])

mean_arterial_bloodpress = []

for x in bloodpress:
    vals = x.split('/')
    map = 0.33* int(vals[0]) + 0.66 * int(vals[1])
    mean_arterial_bloodpress.append(map)

df['Blood Pressure'] = mean_arterial_bloodpress
df.head()

    
```

```python
le = preprocessing.LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])
df['Diet'] = le.fit_transform(df['Diet'])
df['Country'] = le.fit_transform(df['Country'])
df['Continent'] = le.fit_transform(df['Continent'])

```
